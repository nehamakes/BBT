### Children Relationship

```python
def assign_pos_score(x):
    if x == 'Strongly Disagree':
        score = 1
    elif x == 'Disagree':
        score = 2
    elif x == 'Undecided':
        score = 3
    elif x == 'Agree':
        score = 4
    elif x == 'Strongly Agree':
        score = 5
    return score
```

```python
def assign_neg_score(x):
    if x == 'Strongly Disagree':
        score = 5
    elif x == 'Disagree':
        score = 4
    elif x == 'Undecided':
        score = 3
    elif x == 'Agree':
        score = 2
    elif x == 'Strongly Agree':
        score = 1
    return score
```

```python
extrainfo[extrainfo["question"]== "Having child(ren) gives me a more certain and optimistic view for the future."]['answer_value'].apply(lambda x:assign_neg_score(x))
```

```python
def children_rel(df, pos_q_list, neg_q_list):
    def assign_score(x):
        if x.question in pos_q_list:
            return assign_pos_score(x.answer_value)
        elif x.question in neg_q_list:
            return assign_neg_score(x.answer_value)
        else:
            return pd.np.nan
        
    df["score"] = df.apply(lambda x: assign_score(x), axis=1)

    return df
                            
```

```python
pos_q_list = ["Having child(ren) gives me a more certain and optimistic view for the future."
             ,"I am happy in my role as a parent"
             ,"I am satisfied as a parent"
             ,"I enjoy spending time with my child(ren)."
             ,"I feel close to my child(ren)."
             ,"I find my child(ren) enjoyable"
             ,"My child(ren) is an important source of affection for me."
             ,"There is little or nothing I wouldn't do for my child(ren) if it was necessary."
             
             ]
```

```python
neg_q_list = ['Having child(ren) has been a financial burden.'
             ,"Caring for my child(ren) sometimes takes more time and energy than I have to give."
             ,"Having child(ren) has meant having too few choices and too little control over my life."
             ,"Having child(ren) leaves little time and flexibility in my life."
             ,"I feel overwhelmed by the responsibility of being a parent."
             ,"If I had it to do over again, I might decide not to have child(ren)."
             ,"I sometimes worry whether I am doing enough for my child(ren)."
             ,"It is difficult to balance different responsibilities because of my child(ren)."
             ,"The behaviour of my child(ren) is often embarrassing or stressful to me."
             ,"The major source of stress in my life is my child(ren)."]
```

```python
extrainfo_copy = children_rel(extrainfo, pos_q_list, neg_q_list)
```

```python
extrainfo_copy = extrainfo_copy.dropna()
```

```python
extrainfo_copy.head()
```

```python
child_rel_score = pd.DataFrame(extrainfo_copy.groupby("subscriber")['score'].sum())
```

```python
s = pd.merge(s,child_rel_score,on='subscriber',how='left')
```

```python
s = s.rename(columns = {'score':"children relationship score"})
```
